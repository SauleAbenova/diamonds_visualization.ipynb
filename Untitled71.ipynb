```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 로드
diamonds = sns.load_dataset("diamonds")

# 데이터 확인
print(diamonds.head())
```

```python
plt.figure()
sns.countplot(data=diamonds, x="cut")
plt.title("Distribution of Cut")
plt.show()
```

```python
plt.figure()
sns.countplot(data=diamonds, x="color")
plt.title("Distribution of Color")
plt.show()
```

```python
plt.figure()
sns.histplot(data=diamonds, x="depth")
plt.title("Histogram of Depth")
plt.show()
```

```python
plt.figure()
sns.histplot(data=diamonds, x="price")
plt.title("Histogram of Price")
plt.show()
```

```python
plt.figure()
sns.boxplot(data=diamonds, y="depth")
plt.title("Boxplot of Depth")
plt.show()
```

```python
plt.figure()
sns.boxplot(data=diamonds, y="price")
plt.title("Boxplot of Price")
plt.show()
```
